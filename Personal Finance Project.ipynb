# Creating general function to get all data easily

```python
def filter_and_save_data(data, column, keywords, output_csv_filename):
    mask = data[column].str.contains('|'.join(keywords), case=False, na=False)
    filtered_data = data[mask]
    filtered_data.to_csv(output_csv_filename, index=False)
```
